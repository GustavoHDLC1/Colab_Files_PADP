# Limpieza y Transformación de Datos

```python
import pandas as pd
import numpy as np
```

```python
# lectura del csv omitiendo el último registro
df = pd.read_csv('/content/drive/MyDrive/MIAAD/PADP/miaad-nyc-r-s.csv').iloc[:-1]
df.tail(3)
```

```python
print(df.shape)
```

```python
# nombre de las columnas
df.columns.tolist()
```

```python
# cambio en el nombre de la primer columna
df = df.rename(columns={"Unnamed: 0": "INDEX MIIAD"})
df.columns.tolist()
```

```python
# pasamos a minúsculas y reemplazamos espacios por guiones bajos
df.columns = df.columns.str.lower().str.replace(' ','_')
df.columns.tolist()
```

```python
# tipo de dato de cada columna
df.dtypes
```

```python
# sale_price, land_square_feet, gross_square_feet --> float
cols = ["sale_price", "land_square_feet", "gross_square_feet"]
df[cols] = df[cols].replace(" -  ", np.nan).applymap(pd.to_numeric)
```
