## Ejercicio 01. Datos Perdidos.

```python
# Carga el conjunto de datos dirty.csv y verifica que su carga sea correcta.
df = pd.read_csv('/content/drive/MyDrive/MIAAD/PADP/dirtydata.csv')
df.head()
```

```python
# Imprime la cantidad de datos perdidos que tiene el conjunto de datos
df.isnull().sum()
```

```python
# Calcula la media de la columna Calories
mean_cal = df['Calories'].mean()
print(mean_cal)
```

```python
# Calcula de nuevo la media de la columna Calories despu√©s de que has imputado.
mean_cal2 = df['Calories'].mean()
print(mean_cal2)
```

## Ejercicio 02. Duplicidad Parcial Una columna: Contar y Filtrar los registros duplicados en Duration

```python
# Carga el conjunto de datos dirty.csv y verifica que su carga sea correcta.
df = pd.read_csv('/content/drive/MyDrive/MIAAD/PADP/dirtydata.csv')
df.head()
```

```python
# Cuenta la cantidad de registros duplicados utilizando solo la informaci√≥n de columna Duration e imprime
dup_dur = df.duplicated(subset=['Duration'], keep='first').sum()
print(dup_dur)
```

```python
# Utiliza value_counts() e imprime la cantidad de registros que hay por cada valor √∫nico.
df['Duration'].value_counts()
```

```python
# Filtra los datos y obten los registros sin duplicados
df[~df.duplicated(subset=['Duration'], keep=False)]
```

## Ejercicio 05: C√°lculo de la media con duplicados y sin duplicados

```python
# Carga el conjunto de datos dirty.csv y verifica que su carga sea correcta.
df = pd.read_csv('/content/drive/MyDrive/MIAAD/PADP/dirtydata.csv')
df.head()
```

```python
# Calcula e imprime la media de las columnas Pulse y MaxPulse
mean_pul = df['Pulse'].mean()
mean_max = df['Maxpulse'].mean()
print(mean_pul)
print(mean_max)
```

```python
# Identifica los valores duplicados en las columnas Pulse y MaxPulse (Duplicidad Parcial) y elimina los registro repetidos. Qu√©date con las √∫ltimas ocurrencias
df = df[~df.duplicated(subset=['Pulse', 'Maxpulse'], keep='last')]
```

```python
# Utiliza el conjunto anterior sin datos duplicados (recuerda que solo quedaron las √∫ltimas ocurrencias de los datos) y calcula la media de las columnas Pulse y MaxPulse
mean_pul2 = df['Pulse'].mean()
mean_max2 = df['Maxpulse'].mean()
print(mean_pul2)
print(mean_max2)
```
