# Práctica 15: Datos Temporales

## Ejercicio 01:Tipo de Datos Correcto

```python
import pandas as pd
#import datetime as dt
```

```python
# Carga el conjunto de datos airdata.csv y verifica que cada columna tenga el tipo de datos correcto
df = pd.read_csv('/content/drive/MyDrive/MIAAD/PADP/Airdata.csv')
df.dtypes
```

```python
# Si una columna de tipo temporal no está en el tipo de correcto realiza la conversión.
df["DateTime"] = pd.to_datetime(df["DateTime"], format="%m/%d/%Y %H:%M")
```

```python
# Usa dtypes para mostrar que cada columna, principalmente la que contiene datos temporales, está en el tipo de datos correcto
df.dtypes
```

```python
# Imprime las primeras tres líneas
df.head(3)
```

## Ejercicio 02: Extraer componentes de fecha y tiempo

```python
# De la columna 'DateTime' extrae el año, mes, día y hora.
# Crea una nueva columna para cada componente que extraes
df['year'] = df['DateTime'].dt.year
df['month'] = df['DateTime'].dt.month
df['day'] = df['DateTime'].dt.day
df['hour'] = df['DateTime'].dt.hour
```

```python
# Imprime las primeras tres líneas del dataframe que ahora deberá tener 4 nuevas columnas.
df.head(3)
```

## Ejercicio 03: Aritmética - Suma de Intervalos (timedelta)

```python
# Suma 10 días, 7 horas y 15 minutos a cada elemento de la columna 'DateTime'.
# Guarda el resultado en una nueva columna y agrega al dataframe.
df['DateTime2'] = df['DateTime'] + pd.Timedelta(days=10, hours=7, minutes=15)
```

```python
# Imprime la columna 'DateTime' y la columna que resulta de sumar 10 días, 7 horas y 15 minutos.
df[['DateTime','DateTime2']]
```

## Ejercicio 04: Timestamp

```python
# Convierte la columna 'DateTime' la cual tiene un formato de fecha legible a a timestamp Unix.
# Guarda el resultado en una nueva columna
df["timestamp"] = df["DateTime"].apply(lambda x: x.timestamp())
```

```python
# Imprime las primeras tres filas de  la columna 'DateTime'
# y la columna que representa dicha fecha en timestamp Unix
df[['DateTime','timestamp']].head(3)
```

## Ejercicio 05: Filtrado de Fechas

```python
# Utiliza la columna 'DateTime' y Filtra aquellas fechas que son mayores a 2020-11-01
filtro = df['DateTime'] > '2020-11-01'
```

```python
# Imprime el resultado
df[filtro]
```
